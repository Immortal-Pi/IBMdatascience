```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html'
```

```python
data=requests.get(url).text
# data
```

# parsing the HTML content


```python
soup=BeautifulSoup(data,'html5lib')
```

```python
netflixdata=pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close","Volume"])
```

```python

for row in soup.find_all('tbody')[0].find_all('tr'):
    col=row.find_all('td')
    date=col[0].text
    open=col[1].text
    high=col[2].text
    low=col[3].text
    close=col[4].text
    adj=col[5].text
    vol=col[6].text
    netflixdata=pd.concat([netflixdata,pd.DataFrame({'Date':[date],'Open':[open],'High':[high],'Low':[low],'Close':[close],'Adj Close':[adj],'Volume':[vol]})],ignore_index=True)
```

```python
netflixdata
```

# Exercise: use webscraping to extract stock data

```python
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html'
data=requests.get(url).text
# data
```

```python
soup=BeautifulSoup(data,'html5lib')
```

```python
amazon_data=pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close","Volume"])
```

```python
for row in soup.find_all('tbody')[0].find_all('tr'):
    col=row.find_all('td')
    date=col[0].text
    open=col[1].text
    high=col[2].text
    low=col[3].text
    close=col[4].text
    adj=col[5].text
    vol=col[6].text
    data=pd.DataFrame({"Date":[date], "Open":[open], "High":[high], "Low":[low], "Close":[close], "Adj Close":[adj],"Volume":[vol]})
    amazon_data=pd.concat([amazon_data,data],ignore_index=True)
```
