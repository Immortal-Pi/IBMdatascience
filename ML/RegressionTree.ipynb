# Regression Tree

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('datasets/real_estate_data.csv')
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

```python
x=data.drop(columns=['MEDV'])
y=data['MEDV']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=1)
```

# creating regression tree


```python
regression_tree=DecisionTreeRegressor(criterion='friedman_mse')
regression_tree
```

```python
regression_tree.fit(xtrain,ytrain)
```

```python
regression_tree.score(xtest,ytest)
```

```python
pred=regression_tree.predict(xtest)
print(f'prediction: {(pred-ytest).abs().mean()*1000}')
```

```python
regression_tree2=DecisionTreeRegressor(criterion='absolute_error')
```

```python
regression_tree2.fit(xtrain,ytrain)
```

```python
regression_tree2.score(xtest,ytest)
```

```python
pred=regression_tree2.predict(xtest)
print(f'{(pred-ytest).abs().mean()*1000}')
```
