# Visualizing Data using Matplotlib

# Line Pots (Series/Dataframe) <a id="4"></a>

```python
%matplotlib inline
```

```python
haiti=data.loc['Haiti',year]
haiti.head()
```

```python
haiti.plot()
```

```python

haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```

```python
data_immigrants=data.loc[['India','China'],range(1980,2013)]
data_immigrants
```

```python
data_immigrants.plot(kind='line')
```

# the above graph doesnt look right therefore we need to transpose it

```python
data_immigrants=data_immigrants.transpose()
```

```python
data_immigrants.head()
```

```python
data_immigrants.plot()
plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

**Question:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
data_immigration_top5=data.sort_values(by='total',ascending=False,axis=0).head(5)
data_immigration_top5=data_immigration_top5.iloc[:,3:]
data_immigration_top5
```

```python
data_immigration_top5=data_immigration_top5.T
```

```python
data_immigration_top5
```

```python
data_immigration_top5.plot()
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```
