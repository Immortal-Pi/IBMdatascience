# AREA PLOTS

```python
data_immigration_top5.drop('total',axis=0).plot(kind='area',alpha=0.8,stacked=False)
plt.title('Immigration Trend of Top 5 Countries to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

# two types of plotting
**Option 1: Scripting layer (procedural method) - using matplotlib.pyplot as 'plt'**

**Option 2: Artist layer (Object oriented method) - using an `Axes` instance from Matplotlib (preferred)**

```python
# Option 1: This is what we have been using so far
data_immigration_top5.drop('total',axis=0).plot(kind='area', alpha=0.35, figsize=(20, 10)) 
plt.title('Immigration trend of top 5 countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
# option 2: preferred option with more flexibility
ax=data_immigration_top5.drop('total',axis=0).plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
data_immigration_least=data.sort_values(by='total',ascending=True,axis=0).head(5)
data_immigration_least=data_immigration_least.iloc[:,3:]
data_immigration_least=data_immigration_least.T
data_immigration_least.drop('total',axis=0,inplace=True)
```

# Histograms


**Question:** What is the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?

```python
count,bin_edge= np.histogram(data[2013])
```

```python
count, bin_edge
```

```python
data[2013].plot(kind='hist',figsize=(8, 5))
plt.title('Histogram of Immigration from 195 Countries in 2013')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')
```

```python
data[2013].plot(kind='hist',figsize=(8, 5),xticks=bin_edge)
plt.title('Histogram of Immigration from 195 Countries in 2013')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')
```

```python
years = list(range(1980, 2014))
years
```

```python
data_norway=data.loc[['Denmark','Norway','Sweden'],years]
```

```python
data_norway.plot.hist()
```

```python
data_norway.T.plot.hist()
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

Let's make a few modifications to improve the impact and aesthetics of the previous plot:

*   increase the bin size to 15 by passing in `bins` parameter;
*   set transparency to 60% by passing in `alpha` parameter;
*   label the x-axis by passing in `x-label` parameter;
*   change the colors of the plots by passing in `color` parameter.


```python
count,bin_edge=np.histogram(data_norway)
data_norway.T.plot(
    figsize=(10,6),
    kind='hist',
    bins=15,
    alpha=0.6,
    xticks=bin_edge,
     color=['coral', 'darkslateblue', 'mediumseagreen']
)
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
```

# Pie Charts


```python
data_continents=data.groupby('Continent',axis=0).sum()
data_continents.drop(columns=['Region','DevName'],inplace=True)
data_continents
```

```python
data_continents['total'].plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(10,6),
    startangle=90,
    shadow=True,
    labels=None,
    pctdistance=1.12
)
plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels=data_continents.index, loc='upper left') 
```
