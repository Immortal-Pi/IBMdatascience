# dataframe from dictionaries

```python
data2 = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}
df3 = pd.DataFrame(data2)
df3
```

## conditional filtering

```python
df4=df3[df3['Age']>=30]
df4
```

# saving dataframe

```python
df3.to_csv('datasets/newNames.csv')
```

```python
x=music_dataset['Length']
type(x)
```

```python
x=music_dataset[['Length']]
type(x)
```
