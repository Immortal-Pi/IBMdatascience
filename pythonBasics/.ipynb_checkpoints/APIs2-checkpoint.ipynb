```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
URL="https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
tables=pd.read_html(URL)
data=tables[3]
type(data)
```

```python
data.columns = range(data.shape[1])
data
```

```python
data1=data[[0,2]]
data1=data1.iloc[1:11,:]
data1
```

```python
data1.columns=['Country','GDP(Million USD)']
data1['GDP(Million USD)']=data1['GDP(Million USD)'].astype(int)
data1
```

# Exercise 2

```python
data1[['GDP(Million USD)']]=data1[['GDP(Million USD)']]/1000
data1[['GDP(Million USD)']]=np.round(data1[['GDP(Million USD)']],2)
data1
```

```python
data1.rename(columns={'GDP(Million USD)':'GDP(Billion USD)'},inplace=True)
data1
```

```python
data1.to_csv('datasets/Largest_economies.csv')
```
