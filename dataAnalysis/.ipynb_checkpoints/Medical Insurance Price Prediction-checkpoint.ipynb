# Insurance Cost Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from ipywidgets import interact, interactive, fixed, interact_manual
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge
```

# Task 1 : Import the dataset

```python
medi_data=pd.read_csv('datasets/medical_insurance_dataset.csv')
medi_data
```

```python
medi_data.columns=["age", "gender", "bmi", "no_of_children", "smoker", "region", "charges"]
medi_data.head(5)
```

```python
medi_data.replace('?',np.NaN, inplace=True)
```

```python
medi_data.dropna(inplace=True)
medi_data.shape
```

# Task 2 : Data Wrangling

```python
medi_data.info()
```

```python
missing_data=medi_data.isnull()
for col in missing_data.columns.values.tolist():
    print(col)
    print(missing_data[col].value_counts())
    print('')
missing_data
```

```python
is_smoker=medi_data['smoker'].value_counts().idxmax()
medi_data['smoker'].replace(np.nan,is_smoker,inplace=True)

mean_age=medi_data['age'].astype('float').mean(axis=0)
medi_data['age'].replace(np.nan,mean_age,inplace=True)
medi_data[['age','smoker']]=medi_data[['age','smoker']].astype(int)
```

```python
# medi_data[['smoker']]=medi_data[['smoker']].replace({'yes':1,'no':0})
```

```python
medi_data.info()
```

```python
medi_data[['charges']]=np.round(medi_data[['charges']],2)
medi_data.shape
```

```python
medi_data['gender'].value_counts()
```

# Task 4 : Model Development

```python
lm1=LinearRegression()
x_data=medi_data[['smoker']]
y_data=medi_data['charges']
```

```python
lm1.fit(x_data,y_data)
```

```python
lm1.score(x_data,y_data)
```

```python
input=[('scalar',StandardScaler()),('Polynomial',PolynomialFeatures()),('model',LinearRegression())]
```

```python
pipe=Pipeline(input)
pipe
```

```python
z=medi_data.drop(columns='charges')
# z=medi_data[["age", "gender", "bmi", "no_of_children", "smoker", "region"]]
y=medi_data['charges']
```

```python
pipe.fit(z,y)
ypipe=pipe.predict(z)
```

```python
print(r2_score(ypipe,y))
```

# Task 5 : Model Refinement

```python
xtrain,xtest,ytrain,ytest=train_test_split(z,y,test_size=0.2,random_state=1)
```

```python
RR=Ridge(alpha=0.1)
RR.fit(xtrain,ytrain)
yhat=RR.predict(xtest)
print(r2_score(ytest,yhat))
yhat
```

```python
pr=PolynomialFeatures(degree=2)
```

```python
x_train_pr=pr.fit_transform(xtrain)
x_test_pr=pr.fit_transform(xtest)
RR.fit(x_train_pr,ytrain)
y_hat=RR.predict(x_test_pr)
```

```python
print(r2_score(ytest,y_hat))
```

```python
plt.plot(yhat)
plt.plot(ytest)
```
