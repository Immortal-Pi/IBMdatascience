# creating and accessing SQLite database using python



```python
import pandas as pd
import sqlite3 as sql
import seaborn as sns
import matplotlib
```

```python
cnn=sql.connect('instructor.db')
```

```python
cursor=cnn.cursor()
```

```python
cursor.execute('DROP TABLE IF EXISTS INSTRUCTOR')
```

```python
table = """ create table IF NOT EXISTS INSTRUCTOR(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CCODE CHAR(2));"""
```

```python
cursor.execute(table)
 
print("Table is Ready")
```

```python
cursor.execute('''insert into INSTRUCTOR values (1, 'Rav', 'Ahuja', 'TORONTO', 'CA')''')
```

```python
cursor.execute('''insert into INSTRUCTOR values (2, 'Raul', 'Chong', 'Markham', 'CA'), (3, 'Hima', 'Vasudevan', 'Chicago', 'US')''')
```

```python
statement = '''SELECT * FROM INSTRUCTOR'''
cursor.execute(statement)
output=cursor.fetchall()
#we get tuples in a list
type(output[0])
```

```python
statement = '''SELECT * FROM INSTRUCTOR'''
cursor.execute(statement)
output=cursor.fetchmany(3)
#we get tuples in a list
output
```

```python
# Fetch only FNAME from the table
statement = '''SELECT FNAME FROM INSTRUCTOR'''
cursor.execute(statement)
  
print("All the data")
output_column = cursor.fetchall()
for fetch in output_column:
  print(fetch)
```

```python
statement='update instructor set city="moosetown" where fname="Rav"'
cursor.execute(statement)
output
```

```python
statement = '''SELECT * FROM INSTRUCTOR'''
cursor.execute(statement)
  
print("All the data")
output1 = cursor.fetchmany(2)
for row in output1:
  print(row)
```
